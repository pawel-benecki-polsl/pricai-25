```python
import glob
import os
import pandas as pd
import csv 

dfs = dict()

def _split(part: str, typeid):
    kv = part.split("(")
    return kv[0], typeid(kv[1][:-1])


for base in [
    "c:\\experiments\\20230104\\20230101_MuS2_L1",
    "c:\\experiments\\20230104\\20230101_MuS2_CPSNR2",
    "c:\\experiments\\20230104\\20230101_MuS2_LPIPS",
    "c:\\experiments\\20230104\\20230101_MuS2_LPIPS_KFS_20221206",
    "c:\\experiments\\20230117\\mus2_lpips_1vote_kfs_2_votes_human",
    "c:\\experiments\\20230120\\mus2_l1_dotrenowane_lpips_1vote_kfs_2_votes_human",
    "c:\\experiments\\20230120\\mus2_lpips_1vote_kfs_2_votes_human_dotrenowane_l1",
]:
    index_fn = os.path.join(base, "__index.csv")

    with open(index_fn, 'w', newline='') as csvfile: 
        csvwriter = csv.writer(csvfile) 
        csvwriter.writerow(["epoch", "img", "psnr", "lpips", "lpipkfs", "path"]) 

        for f in glob.glob(os.path.join(base, "*_img(???)_*.png")):
            parts = os.path.splitext(os.path.basename(f))[0].split("_")
            item = [
                    _split(parts[0], int)[1],
                    _split(parts[1], int)[1],
                    _split(parts[2], float)[1],
                    _split(parts[3], float)[1],
                    _split(parts[4], float)[1],
                    f
            ]
            csvwriter.writerow(item)

    df = pd.read_csv(index_fn)
    dfs[base] = df
    print(f"Done: {base}")
```
